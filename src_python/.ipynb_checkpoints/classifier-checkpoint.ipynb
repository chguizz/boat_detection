# Classification for boats detection

```python
# Import libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob
import pandas as pd
import random as rd

import tensorflow as tf
import keras
from sklearn.model_selection import train_test_split
np.random.seed(1)
```

## Utils function

```python
def iou(boxA, boxB):
    """
    Compute the IOU metric
    -----------------
    Parameters:
        boxA, boxB are list of corners, top left corner and bottom right corner
        in total 4 coordinates for box
    """
    # Check if there is interseption
    cond1 = boxB[0] >= boxA[0] and boxB[0] <= boxA[2]
    cond2 = boxB[1] >= boxA[1] and boxB[1] <= boxA[3]
    cond3 = boxA[0] >= boxB[0] and boxA[0] <= boxB[2]
    cond4 = boxA[1] >= boxB[1] and boxA[1] <= boxB[3]
    if not ((cond1 and cond2) or (cond3 and cond4)):
        return 0
    
    # determine the (x, y)-coordinates of the intersection rectangle
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    # compute the area of intersection rectangle
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    # compute the area of both the prediction and ground-truth
    # rectangles
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)
    # return the intersection over union value
    return iou
```

```python
def plot_loss_accuracy(history, folder=None):
    fig, ax = plt.subplots(2,1, figsize=(8,10))
    ax[0].plot(history.epoch,history.history['loss'], label="loss", color='k', marker='.', linestyle='--')
    ax[0].plot(history.epoch,history.history['val_loss'], label="val_loss", color='k', marker='.', linestyle='-')
    ax[0].legend()
    ax[0].title.set_text('Binary crossentropy')
    
    ax[1].plot(history.epoch,history.history['accuracy'], label="accuracy", color='k', marker='.', linestyle='--')
    ax[1].plot(history.epoch,history.history['val_accuracy'], label="val_accuracy", color='k', marker='.', linestyle='-')
    ax[1].legend()
    ax[1].title.set_text('Accuracy')
    
    if folder != None:
        # Saving figure
        name = folder.split('/')[-1]
        plt.savefig("../report/figures/"+name+"_plots.jpg")
    
    plt.show()
```

```python
def extract_name(absolute_path):
    i = 0
    for j in range(len(absolute_path)):
        char = absolute_path[j]
        if char == '/' or char == '\\':
            i = j
    name = absolute_path[i+1:-4]
    return name
```

```python
def patch(img, box, resize=True):
    """
    It returns the corresponding patch given a bounding box and an image
    box has structure [xa, ya, xb, yb]
    (xa, ya) top left corner
    (xb, yb) bottom right corner
    """
    xa = box[0]
    ya = box[1]
    xb = box[2]
    yb = box[3]
    
    result = img[ya:yb, xa:xb]
    
    if resize:
        result = cv2.resize(result, (224,224), interpolation = cv2.INTER_LINEAR)
    return result
```

## Classification

Loading ground truth bounding boxes as pandas dataframes

```python
gt_boxes = []
folder = "../dataset/FINAL_DATASET/TRAINING_DATASET/LABELS_TXT"
folder +=  "/*.txt"

filenames = []

for path in glob.glob(folder):
    filenames.append(extract_name(path))
    
    # We use a pandas dataframe for each image to store all bounding boxes
    df = pd.read_csv("../dataset/FINAL_DATASET/TRAINING_DATASET/LABELS_TXT/image0001.txt", sep=';', header=None).iloc[:,:-1]
    series = df.iloc[:,0]
    new_column = []
    for elem in series:
        new_column.append(int(elem[5:]))
    df.iloc[:,0] = new_column
    
    gt_boxes.append(df)
    
print("Number of boats with labels: {}".format(len(gt_boxes)))
```

Example of bounding boxes given an image:

```python
gt_boxes[0]
```

Loading images

```python
folder = "../dataset/FINAL_DATASET/TRAINING_DATASET/IMAGES"
X_boats = []

for name in filenames:
    file = folder + "/" + name + ".png"
    X_boats.append(cv2.cvtColor(cv2.imread(file), cv2.COLOR_BGR2RGB))
    
print("Number of boats loaded: {}".format(len(X_boats)))
```

### ML Dataset creation

Class 1: boats

```python
# Resizing all images
# Bilinear interpolation
image_size = (224,224)
X = []

for i in range(len(X_boats)):
    img = X_boats[i]
    for j in range(gt_boxes[i].shape[0]):
        true_box = gt_boxes[i].iloc[j,:].to_list()
        X.append(patch(img, [true_box[0], true_box[2], true_box[1], true_box[3]]))
```

```python
Y = [1 for elem in X]
```

```python
print("Number of class 1 samples: {}".format(len(X)))
print("Number of labels: {}".format(len(Y)))
```

Class 0: others

```python
MAX_NEGATIVES = 2

count = 0
for n in range(MAX_NEGATIVES):
    for i in range(len(X_boats)):
        img = X_boats[i]
        x = rd.randint(0, img.shape[1] - image_size[0])
        y = rd.randint(0, img.shape[0] - image_size[1])

        box = [x, y, x + image_size[0], y + image_size[1]]

        # If the bounding box chosen randomly doesn't have a IoU > 0.5
        # with any true bounding box, than we consider this a false image
        # Class 0 of no-boats
        check = True
        for j in range(gt_boxes[i].shape[0]):
            true_box = gt_boxes[i].iloc[j,:].to_list()
            if iou(box, [true_box[0], true_box[2], true_box[1], true_box[3]]) > 0.5: 
                check = False
        if check:
            X.append(patch(img, box, False))
            Y.append(0)
            count += 1
```

```python
print("Number of class 0 samples: {}".format(count))
print()
print("Number of TOTAL samples: {}".format(len(X)))
print("Number of labels: {}".format(len(Y)))
```

We convert everything into numpy array objects and split X and Y in training set and validation set.

```python
X = np.array(X)
print("X:")
print(X.shape)
print()
print("Number of samples: {}".format(X.shape[0]))
print("Image shape: ({},{})".format(X.shape[1], X.shape[2]))
print("Number of channels: {}".format(X.shape[3]))

Y = np.array(Y)
print()
print("Y:")
print(Y.shape)
```

```python
# Now we split the dataset into
# 1. training set
# 2. validation set

X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, shuffle=True, random_state=1)
print("Training set: {} and {}".format(X_train.shape, y_train.shape))
print("Validation set: {} and {}".format(X_val.shape, y_val.shape))
```

```python
# Input normalization

# subtraction of the mean and divide everything
# for the standard deviation

for i in range(X_train.shape[0]):
    X_train[i,:,:,:] = tf.image.per_image_standardization(X_train[i,:,:,:])
for i in range(X_val.shape[0]):
    X_val[i,:,:,:] = tf.image.per_image_standardization(X_val[i,:,:,:])
```

## Models

----------------------------------------------------------- Shallow CNN -----------------------------------------------------------

```python
model_shallow = keras.models.Sequential([
    keras.layers.Conv2D(filters=32, kernel_size=[3,3], padding="same", activation="relu", input_shape=X.shape[1:]),
    keras.layers.MaxPool2D(pool_size=[2,2]),
    keras.layers.Flatten(),
    keras.layers.Dense(1, activation="sigmoid")
])
model_shallow.compile(loss="binary_crossentropy",
              optimizer='adam',
              metrics=["accuracy"])
```

```python
model_shallow.summary()
```

```python
history1 = model_shallow.fit(X_train, y_train, epochs=10, batch_size=128, validation_data=(X_val, y_val))
```

```python
plot_loss_accuracy(history1)
```

----------------------------------------------------------- Deep CNN 1 -----------------------------------------------------------

```python
model_deep = keras.models.Sequential([
    keras.layers.Conv2D(filters=32, kernel_size=[3,3], activation='relu', input_shape=X.shape[1:]),
    keras.layers.MaxPool2D(pool_size=[2,2]),
    keras.layers.Conv2D(filters=16, kernel_size=[2,2], activation='relu'),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(1, activation="sigmoid", kernel_regularizer=tf.keras.regularizers.l2(0.01))
  ])
model_deep.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
  
print(model_deep.summary())
```

```python
history2 = model_deep.fit(X_train, y_train, epochs=10, batch_size=128, validation_data=(X_val, y_val))
```

```python
plot_loss_accuracy(history2)
```

----------------------------------------------------------- Deep CNN 2 -----------------------------------------------------------

```python
model_deep2 = keras.models.Sequential([
    keras.layers.Conv2D(filters=32, kernel_size=[3,3], activation='relu', input_shape=X.shape[1:]),
    keras.layers.Conv2D(filters=16, kernel_size=[3,3], activation='relu'),
    keras.layers.MaxPool2D(pool_size=[2,2]),
    keras.layers.Conv2D(filters=8, kernel_size=[2,2], activation='relu'),
    keras.layers.Conv2D(filters=4, kernel_size=[2,2], activation='relu'),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(1, activation="sigmoid", kernel_regularizer=tf.keras.regularizers.l2(0.01))
  ])
model_deep2.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
  
print(model_deep2.summary())
```

```python
history3 = model_deep2.fit(X_train, y_train, epochs=10, batch_size=128, validation_data=(X_val, y_val))
```

```python
plot_loss_accuracy(history3)
```

Summary plot

```python
fig,ax =plt.subplots(1,2,figsize=(16,6))

color1 = "#ff6927"
color2 = "#9e4692"
color3 = "#003f5c"

# *** LOSS ***
# --------------shallow model
ax[0].plot(history1.epoch, history1.history['loss'], label='train sw', color=color1, linestyle='--', marker='.')
ax[0].plot(history1.epoch, history1.history['val_loss'], label='val sw', color=color1, linestyle='-', marker='.')
# --------------deep model 1
ax[0].plot(history2.epoch, history2.history['loss'], label='train dp1', color=color2, linestyle='--', marker='.')
ax[0].plot(history2.epoch, history2.history['val_loss'], label='val dp1', color=color2, linestyle='-', marker='.')
# --------------deep model 2
ax[0].plot(history3.epoch, history3.history['loss'], label='train dp2', color=color3, linestyle='--', marker='.')
ax[0].plot(history3.epoch, history3.history['val_loss'], label='val dp2', color=color3, linestyle='-', marker='.')

ax[0].legend()
ax[0].set_title('loss')

# *** ACCURACY ***
# --------------shallow model
ax[1].plot(history1.epoch, history1.history['accuracy'], label='train sw', color=color1, linestyle='--', marker='.')
ax[1].plot(history1.epoch, history1.history['val_accuracy'], label='val sw', color=color1, linestyle='-', marker='.')
# --------------deep model 1
ax[1].plot(history2.epoch, history2.history['accuracy'], label='train dp1', color=color2, linestyle='--', marker='.')
ax[1].plot(history2.epoch, history2.history['val_accuracy'], label='val dp1', color=color2, linestyle='-', marker='.')
# --------------deep model 2
ax[1].plot(history3.epoch, history3.history['accuracy'], label='train dp2', color=color3, linestyle='--', marker='.')
ax[1].plot(history3.epoch, history3.history['val_accuracy'], label='val dp2', color=color3, linestyle='-', marker='.')

ax[1].legend()
ax[1].set_title('accuracy')

plt.savefig("../report/figures/CNNs.png")

plt.show()
```
